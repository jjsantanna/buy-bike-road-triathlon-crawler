```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://buycycle.com/en/shop/bike-types/renn-gravel/categories/triathlon/frame-sizes/m,l/max-price/1600")
```

```python
bikes = driver.find_elements_by_xpath('//div[@class="shop-product-item"]')
b=[]
for bike in bikes:
    b.append(bike.text)
```

```python
links = driver.find_elements_by_xpath('//div[@class="shop-product-item"]/div/a')
l =[]
for link in links:
    l.append(link.get_attribute("href"))
```

```python
df = pd.DataFrame(b)
```

```python
df['links']= pd.DataFrame(l)
```

```python
df['year'] = df[0].apply(lambda x: x.split('\n • \n')[0])
```

```python
df['frame_size'] = df[0].apply(lambda x: x.split('\n • \n')[1].replace(' cm',''))
```

```python
df['set_type'] = df[0].apply(lambda x: x.split('\n • \n')[2].split('\n')[0])
```

```python
df['frame_brand'] = df[0].apply(lambda x: x.split('\n • \n')[2].split('\n')[1])
```

```python
df['frame_model'] =df[0].apply(lambda x: x.split('\n • \n')[2].split('\n')[2])
```

```python
df['price_original'] =df[0].apply(lambda x: x.split('\n • \n')[2].split('\n')[4].replace('€',''))
```

```python
df['price'] =df[0].apply(lambda x: x.split('\n • \n')[2].split('\n')[5].replace('€',''))
```

shimano Tourney>> Claris >> Sora >> Tiagra >> 105 (R7000, R7020) >> Ultegra (R8000, R8020, R8050, R8070) >> Dura-Race (*Di2)

```python
type_set_weight={'Shimano Dura Ace Di2':1,
                 'Shimano Dura Ace':0.9,
                 'SRAM Red':0.9,
                 'Shimano Ultegra Di2':0.8,
                 'Shimano Ultegra':0.7,
                 'Shimano 105': 0.6}
```

```python
df['type_set_weight'] = df.replace({"set_type": type_set_weight})['set_type']
```
